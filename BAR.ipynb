## Total Contribution in a set

```python
"""Assuming your DataFrame is named 'df' + add a column "contribution" 
where adds all values of box (attack+block+serve+set+dig+receive-error)"""

# Create a new column 'cont_in_set' and initialize it with 0
df['cont_in_set'] = 0

# Iterate over the rows and calculate the total contribution for each set
total_contribution = 0
prev_set_info = None
set_indices = []

for index, row in df.iterrows():
    curr_set_info = (row['date'], row['set'], row['team'])
    
    if prev_set_info is None or any(curr != prev for curr, prev in zip(curr_set_info, prev_set_info)):
        # New set, update the previous set's contribution
        if prev_set_info is not None:
            for set_index in set_indices:
                df.loc[set_index, 'cont_in_set'] = total_contribution
        
        total_contribution = 0  # Reset the total contribution
        set_indices = [index]
    else:
        set_indices.append(index)
    
    total_contribution += row['contribution']  # Add to the total contribution
    
    prev_set_info = curr_set_info

# Update the final set's contribution
for set_index in set_indices:
    df.loc[set_index, 'cont_in_set'] = total_contribution

# Sort the DataFrame back to the original order
df.sort_index(inplace=True)
```

```python
df.head(50)
```

### # Add a new column 'true_cont' to calculate the percentage of 'contribution' to 'cont_in_set'

```python
df['true_cont'] = df['contribution'] / df['cont_in_set']
df.head(50)
```

### Double check if the contribtution percentage (true_cont) adds up to 1 in all sets

```python
# Group the DataFrame by 'date', 'set', and 'team' and check if the sum of 'true_cont' within each group is approximately 1
grouped = df.groupby(['date', 'set', 'team'])
group_sums = grouped['true_cont'].sum()
group_sums_equal_1 = group_sums.abs().sub(1).lt(1e-10)  # Checking if the sum is approximately 1 with a tolerance of 1e-10

# Add a new column 'true_cont_sum_equals_1' to indicate if the sum of 'true_cont' for each set equals 1
df['true_cont_sum_equals_1'] = df.apply(lambda row: group_sums_equal_1[(row['date'], row['set'], row['team'])], axis=1)
df.head(50)
```

```python
# Save df as Excel
df.to_excel('boxscore2122.xlsx', index=False)
```
